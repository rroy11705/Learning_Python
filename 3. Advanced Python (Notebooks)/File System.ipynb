```python
import pickle

def storeData():
    pickle_dataset_1 = {
        "key": "john",
        "name": "fred",
        "age": 21,
        "pay": 400000
    }
    pickle_dataset_2 = {
        "key": "jay",
        "name": "jay",
        "age": 50,
        "pay": 500000
    }
    
    db = {}
    db['john'] = pickle_dataset_1
    db['jay'] = pickle_dataset_2
    dbfile = open("example-pickle", "wb")
    pickle.dump(db, dbfile)
    dbfile.close()
    
    
def loadFile():
    dbfile = open("example-pickle", "rb")
    db = pickle.load(dbfile)
    for key in db:
        print(key, '==>', db[key])
    dbfile.close()
    
if __name__ == '__main__':
    storeData()
    loadFile()
```

```python
import pickle

def storeData():
    pickle_dataset_1 = ["rahul","roy",21,400000]
    pickle_dataset_2 = ["rajesh","paul",30,500000]
    pickle_dataset_3 = ["prakash","jha",30,500000]
    db = [pickle_dataset_1, pickle_dataset_2, pickle_dataset_3]
    dbfile = open("example-pickle", "wb")
    pickle.dump(db, dbfile)
    dbfile.close()
    
    
def loadFile():
    dbfile = open("example-pickle", "rb")
    db = pickle.load(dbfile)
    for data in db:
        print(data)
    dbfile.close()
    
if __name__ == '__main__':
    storeData()
    loadFile()
```
